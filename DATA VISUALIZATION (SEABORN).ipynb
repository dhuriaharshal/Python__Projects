```python
# Seaborn is a statistical plotting library.
# It has beautiful default styles.
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Distplot: It allows us to show the distribution of a univariate set of observations.
# It has histogram and kde.

sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
# Jointplot: To show bivariate set of observations.

# kde allows to affect what's actually going inside of the jointplot.
# Hexagon: More the points, darker the area and vice-versa.
# kde: shows the density where the points matchup the most.


# sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
# sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')

sns.jointplot(x='total_bill',y='tip',data=tips)
```

 Pair-plot: Plot pairwise relationships across an entire dataframe atleast for
 numerical columns.
 
 It also supports a color hue argument for categorical columns.
 
 Hue argument colors the data points based off the category we pass in.

```python
# sns.pairplot(tips,hue='sex',palette='coolwarm')

sns.pairplot(tips,hue='sex')
```

```python
# Rugplot: It draws a dash mark for every point on this uniform or unique variant distribution

sns.rugplot(tips['total_bill'])
```

```python
# kdeplot: without any distribution of the bars.

sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
import seaborn as sns
%matplotlib inline
sns.load_dataset('tips')
tips.head()
```

Barplot: Allows to aggregate the categorical data based off of some function and by default that's the mean (Visualisation of group by action)

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)

# By default it shows the average mean of total_bill per categorical column value.
```

```python
# Countplot: Here, estimator is explicitly counting the number of occurences.  
# Thus, we only pass the x values.

sns.countplot(x='sex',data=tips)
```

```python
# Box plot: To show the distribution of the categorical data. Also known as the whisker plot.
# Points outside the whiskers are the outliers.
# We can further split by passing in the hue argument.

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# Violin Plot:

sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

###### Strip plot: Just a scatter plot based off of the category. (ie. where one of the variable is categorical). One problem with this plot is that it can't really tell how many points are stacked on the top of each other, We can use jitter=True to seperate these stacked points.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
# Swarm plot: Basically combining strip plot ie. scatter plot and violen plot. 
# Here points are adjusted so that they don't overlap.
# Problem is here that sometimes they don't actually scale that well to very large numbers both in
# terms of ability to show all the points. And also requires a lot of computation to arrange them.

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')

# Combining both of them will show both the data points and kde on top of it.
```

```python
# Factor plot:

sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## Matrix Plots

```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
# To work with the heat map, we first need to convert the data into matrix form.

tc= tips.corr()
```

```python
tc
```

```python
# Heat Map: Colors up the values based on the gradient scale.

sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
# We can also convert the data into matrix form with the pivot_table.

fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

```python
# Clustermap: It will use the hierarchial clustering.
# It tries to cluster rows and columns together based off their similarity.

sns.clustermap(fp,standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# With PairGrid we can plot specific types we want to diagonal upper or lower.
g= sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
tips.head()
```

```python
# FacetGrid
# First passing the row and col or variable we'd like to seperate
# When mapping we have to put in another argument to display.

s= sns.FacetGrid(data=tips,col='time',row='smoker')
# s.map(sns.distplot,'total_bill')

# If we want to pass in more than one argument, we can use scatter plot.
# As distplot takes up only one argument.

s.map(sns.scatterplot,'total_bill','tip')
```
